```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 4.0
b = 5.0
sigma = 10.0

print (b/(a + b))
w1 = lambda rho: (a/2.0)*(rho/(a + b*np.exp(-sigma*rho**2)) + (rho-1.0)/(a + b*np.exp(-sigma*(rho - 1)**2)))
# w1 = lambda rho: (a)*(1.0/(a + b*np.exp(-sigma*rho**2)) + 1.0/(a + b*np.exp(-sigma*(rho - 1)**2)))
w2 = lambda rho: rho*(rho - 1)*(2*rho - 1)

rhox = np.linspace(-1, 2, 100)

fig, ax = plt.subplots(figsize = (12,8))
ax.plot(rhox, w1(rhox))
ax.plot(rhox, w2(rhox))
ax.axis([-1, 2, -1, 1])
plt.show()
```
