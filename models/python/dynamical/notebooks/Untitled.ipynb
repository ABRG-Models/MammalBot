```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.rand(100)
y = 1*x + np.random.rand(100)*0.1
# y = 5*x + np.random.rand(100) + np.array([2.0 if i > 50 else -2 for i in range(100)])

print(np.corrcoef(x, y))
plt.plot(x, y, '.')
plt.show()
```

```python
L = 2.0
p = L/2.0
n = 1.0
f = lambda x: 2.0/(np.sqrt(x*(L - x))*2.0*np.pi)
g = lambda x: np.sqrt(2.0/p)*np.cos(n*np.pi*x/p)*np.sqrt(2.0/p)*np.cos(n*np.pi*(x - p)/p)
x = np.random.random(1000)*L

plt.figure()
plt.hist(f(x), bins = np.arange(0, 3, 0.02))
plt.title('pdf for the boundaries')
plt.xlabel('Correlation coefficient')
plt.figure()
plt.hist(g(x), bins = np.arange(-3, 3, 0.02))
plt.title('pdf for the cookie cutter case')
plt.xlabel('Correlation coefficient')

plt.figure()
x = np.linspace(0, L, 100)
plt.plot(x, f(x))
plt.title('Correlation function')
plt.xlabel('p')
plt.ylabel('f(p)')
plt.show()
```
